#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(string):
    vowels = ["a", "e", "i", "o", "u"]
    count = 0
    for char in string:
        if char.lower() in vowels:
            count += 1
    return count
```

```python
def test_count_vowels():
    assert count_vowels("hello") == 2
    assert count_vowels("world") == 1
  
test_count_vowels()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(string):
    char_count = {}
    for char in string:
        char_count[char] = char_count.get(char, 0) + 1
        if char_count[char] > 1:
            return False
    return True
```

```python
def test_is_unique():
    assert is_unique("hello") == False
    assert is_unique("world") == True

test_is_unique()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_bits(num):
    count = 0
    while num > 0:
        if num & 1:
            count += 1
        num >>= 1
    return count
```

```python
def test_count_bits():
    assert count_bits(0) == 0
    assert count_bits(5) == 2
    assert count_bits(13) == 3
 
test_count_bits()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def magic(n):
    count = 0

    while n >= 10:
        product = 1
        while n > 0:
            digit = n % 10
            product *= digit
            n //= 10
        n = product
        count += 1

    return count

def test_magic():
    assert magic(39) == 3
    assert magic(4) == 0
    assert magic(999) == 4

    pass
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Vectors must have the same length")

    squared_diffs = [(p - t) ** 2 for p, t in zip(pred, true)]
    mean_squared_diff = sum(squared_diffs) / len(pred)
    return mean_squared_diff

def test_mse():
    pred = [1, 2, 3, 4, 5]
    true = [2, 4, 6, 8, 10]
    assert mse(pred, true) == 5.0

    pred = [0, 0, 0, 0]
    true = [1, 2, 3, 4]
    assert mse(pred, true) == 5.0

    pred = [1, 2, 3]
    true = [4, 5, 6]
    try:
        mse(pred, true)
    except ValueError as e:
        assert str(e) == "Векторы должны иметь одинаковую длину"
    else:
        raise AssertionError("Expected ValueError")

    pass
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def factorization(n):
    if n <= 1:
        return ""

    factors = []
    divisor = 2
    while n > 1:
        if n % divisor == 0:
            count = 0
            while n % divisor == 0:
                count += 1
                n //= divisor
            factors.append((divisor, count))
        divisor += 1

    return '(' + ')('.join([f'{factor}**{count}' if count > 1 else str(factor) for factor, count in factors]) + ')'

def test_factorization():
    assert factorization(1) == ""
    assert factorization(2) == "(2)"
    assert factorization(25) == "(5**2)"
    assert factorization(100) == "(2**2)(5**2)"
    assert factorization(86240) == "(2**5)(5)(7**2)(11)"
    assert factorization(5040) == "(2**4)(3**2)(5)(7)"

    pass
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(number):
    total = 0
    k = 0
    while total < number:
        k += 1
        total += k**2
    if total == number:
        return k
    else:
        return "It is impossible"

def test_pyramid():
    assert pyramid(1) == 1
    assert pyramid(5) == "It is impossible"
    assert pyramid(30) == 4
    assert pyramid(55) == 5
    assert pyramid(100) == "It is impossible"

    pass
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced(num):
    digits = [int(x) for x in str(num)]
    length = len(digits)
    midpoint = length // 2

    if length % 2 == 0:
        left_sum = sum(digits[:midpoint])
        right_sum = sum(digits[midpoint:])
    else:
        left_sum = sum(digits[:midpoint])
        right_sum = sum(digits[midpoint+1:])

    return left_sum == right_sum


def test_is_balanced():
    assert is_balanced(12321) == True
    assert is_balanced(123456) == False
    assert is_balanced(123321) == True
    assert is_balanced(123456789) == False
    assert is_balanced(1234321) == True

    pass
```
